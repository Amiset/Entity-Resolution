```python
import spacy
nlp = spacy.load('en_core_web_sm')
nlp
```

```python
#reading a basic text file
file_name = 'example.txt'
example_file_text = open(file_name).read()

#making an nlp object from the file
example_file_doc = nlp(example_file_text)
# Extract tokens for the given doc
print ([token.text for token in example_file_doc])
```

```python
#print tokens and thier frequency by iterating on the Doc object

for token in example_file_doc:
    print (token, token.idx)
```

```python
#printing stop words in spacy core
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
len(spacy_stopwords)
for stop_word in list(spacy_stopwords)[:12]:
    print(stop_word)
    
print ("---------------------")

for token in example_file_doc:
    if not token.is_stop:
        print(token)
```

```python
about_no_stopword_doc = [token for token in example_file_doc if not token.is_stop]
print (about_no_stopword_doc)
```

```python
#lemmatization - reducing inflected forms of a word while still ensuring that the reduced form belongs to the language
#organizes, organized and organizing are all forms of organize
for token in about_no_stopword_doc:
    print (token, token.lemma_)
```

```python
# taking the text file again and doing a word analysis

from collections import Counter

complete_text = example_file_text
complete_doc = nlp(complete_text)

# Remove stop words and punctuation symbols

words = [token.text for token in complete_doc if not token.is_stop and not token.is_punct]
word_freq = Counter(words)

# 5 commonly occurring words with their frequencies
common_words = word_freq.most_common(5)
print (common_words)
print("--------------------------------------------------")

# Unique words
unique_words = [word for (word, freq) in word_freq.items() if freq == 1]
print (unique_words)
```

```python
from spacy import displacy
about_interest_text = example_file_text
about_interest_doc = nlp(about_interest_text)
displacy.serve(about_interest_doc, style='dep')

#displacy.render(about_interest_doc, style='dep', jupyter=True)
```

```python
#Named Entity Recognition (NER) is the process of locating named entities in unstructured text and then classifying 
#them into pre-defined categories, such as person names, organizations, locations, monetary values, percentages, 
#time expressions, etc

home_bound_text = ("Govind grasped tightly to the top of the seat in front of him to prevent himself from being jostled off his place, his knuckles white due to his tight grip.The bus made headway with the driver trying to keep it steady on the narrow, partly unpaved road. He shook his head, disgruntled and murmured beneath his breath, Kuch nahin badla.Nothing has changed")
home_bound_doc = nlp(home_bound_text)
for ent in home_bound_doc.ents:
    print(ent.text, ent.start_char, ent.end_char,ent.label_, spacy.explain(ent.label_))
```
